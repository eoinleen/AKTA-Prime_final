```python
"""
Scatter Plot Analysis Script - Three Separate Plots Version
============================================================

Generates three plots:
1) Only 1_* entries (red)
2) Only 3_* entries (blue)
3) All entries combined (1_* = red, 3_* = blue)
"""

# ============================================================================
# *** IMPORTANT: SET MODE HERE ***
# ============================================================================
MODE = "batch"   # Options: "single" or "batch"
#
# "single" = Upload one file when prompted (good for Colab)
# "batch"  = Process all .txt files in BATCH_DIRECTORY
# ============================================================================

import os
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
from zipfile import ZipFile
from io import BytesIO

# ============================================================================
# USER OPTIONS
# ============================================================================

DATA_START_ROW = None
X_COLUMN = 0   # First column (target_aligned_rmsd)
Y_COLUMN = 1   # Second column (binder_aligned_rmsd)

# Batch directory (only used if MODE = "batch")
BATCH_DIRECTORY = "/content/drive/MyDrive/scatter_plot_data/20250930_MPNN_evo"

# Prefix configuration
PREFIXES = {
    "1_": {"color": "red", "label": "1_* samples"},
    "3_": {"color": "blue", "label": "3_* samples"}
}

# ============================================================================
# HELPERS
# ============================================================================

def detect_data_start(lines):
    """Find first row with two numeric values."""
    for i, line in enumerate(lines):
        if line.strip().startswith('#'):
            continue
        parts = line.strip().replace(",", " ").split()
        if len(parts) < 2:
            continue
        try:
            float(parts[0])
            float(parts[1])
            return i
        except ValueError:
            continue
    return len(lines)

def parse_column_headers(lines):
    """Extract column names from header lines starting with '#'."""
    headers = {}
    for line in lines:
        if line.startswith("# Column"):
            parts = line.split(":")
            if len(parts) >= 2:
                try:
                    col_index = int(parts[0].split()[2]) - 1
                    col_name = parts[1].strip().split('(')[0].strip()
                    col_name = col_name.replace('*', '').strip()
                    headers[col_index] = col_name
                except (ValueError, IndexError):
                    continue
    return headers

def load_and_parse_file(filepath_or_content, is_content=False):
    """Load data from file or string, auto-detecting format."""
    if is_content:
        lines = filepath_or_content.strip().split("\n")
        filename = "uploaded_file"
    else:
        with open(filepath_or_content, "r") as f:
            lines = f.readlines()
        filename = os.path.basename(filepath_or_content)

    headers = parse_column_headers(lines)
    start_row = DATA_START_ROW if DATA_START_ROW is not None else detect_data_start(lines)

    data_lines = [l for i, l in enumerate(lines)
                  if i >= start_row and not l.strip().startswith("#") and l.strip()]

    if not data_lines:
        print(f"‚ö†Ô∏è  No numeric data found in {filename}, skipping")
        return None

    x_idx = X_COLUMN
    y_idx = Y_COLUMN

    x_vals, y_vals, descs = [], [], []
    for line in data_lines:
        parts = line.strip().replace(",", " ").split()
        try:
            x_vals.append(float(parts[x_idx]))
            y_vals.append(float(parts[y_idx]))
            descs.append(parts[-1] if len(parts) > 2 else "")
        except (ValueError, IndexError):
            continue

    if not x_vals:
        print(f"‚ö†Ô∏è  No valid rows in {filename}")
        return None

    default_xlabel = headers.get(x_idx, "target_aligned_rmsd")
    default_ylabel = headers.get(y_idx, "binder_aligned_rmsd")

    return {
        "filename": filename,
        "x": np.array(x_vals),
        "y": np.array(y_vals),
        "desc": descs,
        "xlabel": default_xlabel,
        "ylabel": default_ylabel
    }

def compute_stats(x, y):
    """Calculate statistics for x and y arrays."""
    if len(x) == 0:
        return None
    corr, pval = pearsonr(x, y)
    return {
        "n_points": len(x),
        "x_min": np.min(x), "x_max": np.max(x),
        "x_mean": np.mean(x), "x_std": np.std(x),
        "y_min": np.min(y), "y_max": np.max(y),
        "y_mean": np.mean(y), "y_std": np.std(y),
        "corr": corr, "pval": pval
    }

def plot_three_versions(dataset, prefixes):
    """Generate three plots: 1_* only, 3_* only, and combined."""

    # Create list to store figure paths
    saved_figures = []
    base_filename = dataset["filename"].replace(".txt", "")

    # Separate data by prefix
    prefix_data = {}
    for prefix in prefixes.keys():
        mask = np.array([d.startswith(prefix) for d in dataset["desc"]])
        if any(mask):
            prefix_data[prefix] = {
                "x": dataset["x"][mask],
                "y": dataset["y"][mask],
                "color": prefixes[prefix]["color"],
                "label": prefixes[prefix]["label"]
            }

    # ========================================================================
    # PLOT 1: Only 1_* entries
    # ========================================================================
    if "1_" in prefix_data:
        data_1 = prefix_data["1_"]
        stats_1 = compute_stats(data_1["x"], data_1["y"])

        fig = plt.figure(figsize=(8, 6))
        plt.scatter(data_1["x"], data_1["y"],
                   color=data_1["color"], alpha=0.5, s=10)
        plt.xlabel(dataset["xlabel"], fontsize=12, fontweight='bold')
        plt.ylabel(dataset["ylabel"], fontsize=12, fontweight='bold')
        plt.title(f"{dataset['filename']} - 1_* entries only\n" +
                 f"Pearson r={stats_1['corr']:.3f}, p={stats_1['pval']:.2e}, n={stats_1['n_points']}",
                 fontsize=11)
        plt.grid(True, linestyle="--", alpha=0.3)
        plt.tight_layout()

        # Save figure
        filename_1 = f"{base_filename}_1star_only.png"
        plt.savefig(filename_1, dpi=300, bbox_inches='tight')
        saved_figures.append(filename_1)
        plt.show()

        print(f"\nüìä Statistics for 1_* entries:")
        print(f"  Points: {stats_1['n_points']}")
        print(f"  X ({dataset['xlabel']}): {stats_1['x_mean']:.3f} ¬± {stats_1['x_std']:.3f}")
        print(f"  Y ({dataset['ylabel']}): {stats_1['y_mean']:.3f} ¬± {stats_1['y_std']:.3f}")
        print(f"  Correlation: r={stats_1['corr']:.3f}, p={stats_1['pval']:.2e}")

    # ========================================================================
    # PLOT 2: Only 3_* entries
    # ========================================================================
    if "3_" in prefix_data:
        data_3 = prefix_data["3_"]
        stats_3 = compute_stats(data_3["x"], data_3["y"])

        fig = plt.figure(figsize=(8, 6))
        plt.scatter(data_3["x"], data_3["y"],
                   color=data_3["color"], alpha=0.5, s=10)
        plt.xlabel(dataset["xlabel"], fontsize=12, fontweight='bold')
        plt.ylabel(dataset["ylabel"], fontsize=12, fontweight='bold')
        plt.title(f"{dataset['filename']} - 3_* entries only\n" +
                 f"Pearson r={stats_3['corr']:.3f}, p={stats_3['pval']:.2e}, n={stats_3['n_points']}",
                 fontsize=11)
        plt.grid(True, linestyle="--", alpha=0.3)
        plt.tight_layout()

        # Save figure
        filename_3 = f"{base_filename}_3star_only.png"
        plt.savefig(filename_3, dpi=300, bbox_inches='tight')
        saved_figures.append(filename_3)
        plt.show()

        print(f"\nüìä Statistics for 3_* entries:")
        print(f"  Points: {stats_3['n_points']}")
        print(f"  X ({dataset['xlabel']}): {stats_3['x_mean']:.3f} ¬± {stats_3['x_std']:.3f}")
        print(f"  Y ({dataset['ylabel']}): {stats_3['y_mean']:.3f} ¬± {stats_3['y_std']:.3f}")
        print(f"  Correlation: r={stats_3['corr']:.3f}, p={stats_3['pval']:.2e}")

    # ========================================================================
    # PLOT 3: All entries combined with color coding
    # ========================================================================
    stats_all = compute_stats(dataset["x"], dataset["y"])

    fig = plt.figure(figsize=(8, 6))

    # Plot each prefix group
    for prefix, data in prefix_data.items():
        plt.scatter(data["x"], data["y"],
                   color=data["color"], label=data["label"],
                   alpha=0.5, s=10)

    plt.xlabel(dataset["xlabel"], fontsize=12, fontweight='bold')
    plt.ylabel(dataset["ylabel"], fontsize=12, fontweight='bold')
    plt.title(f"{dataset['filename']} - All entries\n" +
             f"Pearson r={stats_all['corr']:.3f}, p={stats_all['pval']:.2e}, n={stats_all['n_points']}",
             fontsize=11)
    plt.legend(loc='best')
    plt.grid(True, linestyle="--", alpha=0.3)
    plt.tight_layout()

    # Save figure
    filename_all = f"{base_filename}_combined.png"
    plt.savefig(filename_all, dpi=300, bbox_inches='tight')
    saved_figures.append(filename_all)
    plt.show()

    print(f"\nüìä Statistics for ALL entries:")
    print(f"  Points: {stats_all['n_points']}")
    print(f"  X ({dataset['xlabel']}): {stats_all['x_mean']:.3f} ¬± {stats_all['x_std']:.3f}")
    print(f"  Y ({dataset['ylabel']}): {stats_all['y_mean']:.3f} ¬± {stats_all['y_std']:.3f}")
    print(f"  Correlation: r={stats_all['corr']:.3f}, p={stats_all['pval']:.2e}")

    # Return list of saved figures for zipping
    return saved_figures

# ============================================================================
# MAIN
# ============================================================================

def run_single_mode():
    print("üìÑ Single file mode (upload in Colab)...")
    from google.colab import files
    uploaded = files.upload()
    if not uploaded:
        return

    all_figures = []

    for fn, content in uploaded.items():
        dataset = load_and_parse_file(content.decode("utf-8"), is_content=True)
        if dataset:
            figures = plot_three_versions(dataset, PREFIXES)
            all_figures.extend(figures)

    # Create and download zip file
    if all_figures:
        create_and_download_zip(all_figures, "scatter_plots.zip")

def run_batch_mode():
    if not os.path.exists(BATCH_DIRECTORY):
        print(f"‚ö†Ô∏è Directory {BATCH_DIRECTORY} not found")
        return

    txt_files = [os.path.join(BATCH_DIRECTORY, f)
                 for f in os.listdir(BATCH_DIRECTORY) if f.endswith(".txt")]

    if not txt_files:
        print("‚ö†Ô∏è No .txt files in batch dir")
        return

    print(f"üìÅ Found {len(txt_files)} .txt files\n")

    all_figures = []

    for filepath in txt_files:
        dataset = load_and_parse_file(filepath)
        if dataset:
            print(f"\n{'='*70}")
            print(f"Processing: {dataset['filename']}")
            print('='*70)
            figures = plot_three_versions(dataset, PREFIXES)
            all_figures.extend(figures)

    # Create and download zip file
    if all_figures:
        create_and_download_zip(all_figures, "scatter_plots_batch.zip")

def create_and_download_zip(figure_paths, zip_filename):
    """Create a zip file with all figures and trigger download."""
    print(f"\nüì¶ Creating zip file with {len(figure_paths)} figures...")

    with ZipFile(zip_filename, 'w') as zipf:
        for fig_path in figure_paths:
            if os.path.exists(fig_path):
                zipf.write(fig_path, os.path.basename(fig_path))
                print(f"  ‚úì Added: {os.path.basename(fig_path)}")

    print(f"\n‚úÖ Zip file created: {zip_filename}")

    # Trigger download in Colab
    try:
        from google.colab import files
        files.download(zip_filename)
        print(f"‚¨áÔ∏è  Download started for {zip_filename}")
    except ImportError:
        print(f"üíæ Zip file saved locally: {zip_filename}")

    # Clean up individual PNG files
    for fig_path in figure_paths:
        if os.path.exists(fig_path):
            os.remove(fig_path)

if __name__ == "__main__":
    print(f"üîß Running in {MODE.upper()} mode...\n")

    if MODE == "single":
        run_single_mode()
    else:
        run_batch_mode()
```
